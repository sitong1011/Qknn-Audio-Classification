## Classical KNN

```python
test_normalized = sklearn.preprocessing.normalize(testdata)
print(test_normalized)
```

```python
train_normalized = sklearn.preprocessing.normalize(traindata)
print(train_normalized)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(train_normalized, trainlabels)
```

```python
testlabels=[0,0,1,1]
```

```python
from sklearn.metrics import accuracy_score
correct = np.count_nonzero((clf.predict(test_normalized)==testlabels)==True)
print ("Accuracy is: %.3f" %(correct/len(test_normalized)))
#accuracy_score(y_test, clf.predict(X_test))
```

```python
test_result=clf.predict(testdata)
```
