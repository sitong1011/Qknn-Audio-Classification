## Get The Sample Sound Signal and Resample

```python
# Get x values of the sine wave

time        = np.arange(0, 10, 0.1);
# Amplitude of the sine wave is sine of a variable like time
amplitude   = np.sin(time)
# Plot a sine wave using time and amplitude obtained for the sine wave

plot.plot(time, amplitude)
# Give a title for the sine wave plot

plot.title('Sine wave')
# Give x axis label for the sine wave plot

plot.xlabel('Time')
# Give y axis label for the sine wave plot

plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

Fs = 44100
f = 440
sample = 200
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
xnew = np.linspace(0, 200, 8, endpoint=False)
#xnew = np.arange(0,200,20)
fnew = signal.resample(y, 8)
fig, ax=plt.subplots()
plt.xlabel('Sample(n)')
plt.ylabel('Amplitude')
plt.plot(x, y, '-',xnew, fnew,'r*')
plt.legend(['440Hz wave', 'Resampled'], loc='best')
plt.show()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.spines['left'].set_linewidth(2)
```

## QRDA

## Amplitude Encoding

```python
import matplotlib.pyplot as plt
import numpy as np


Fs = 44100
f = 440
sample = 200
x = np.arange(sample)
fig, ax=plt.subplots()
y = np.sin(2 * np.pi * f * x / Fs)
xnew = np.linspace(0, 200, 8, endpoint=False)
fnew = signal.resample(y, 8)
ampls = fnew / np.linalg.norm(fnew)
print(ampls)
plt.grid(linestyle='-.')
plt.plot(xnew, ampls,'bo')
newx_ticks = np.linspace(0, 200, 9)
plt.xticks(newx_ticks)
newy_ticks = np.linspace(-1, 1, 9)
plt.yticks(newy_ticks)
plt.xticks([0, 25, 50, 75, 100,125,150,175],['$000$', '$001$', '$010$', '$011$', '$100$', '$101$', '$110$', '$111$'])
plt.legend(['Amplitude Encoding'], loc='best')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plot.axhline(y=0, color='k',linewidth=2)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.plot([xnew[1],xnew[1]], [ampls[1],0], color='r', linewidth=1, alpha=0.6)
ax.plot([xnew[3],xnew[3]], [ampls[3],0], color='r', linewidth=1, alpha=0.6)
ax.plot([xnew[5],xnew[5]], [ampls[5],0], color='r', linewidth=1, alpha=0.6)
ax.plot([xnew[7],xnew[7]], [ampls[7],0], color='r', linewidth=1, alpha=0.6)

plt.show()
```
